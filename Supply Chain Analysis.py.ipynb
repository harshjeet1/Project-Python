```python
import pandas as pd
from matplotlib import pyplot as plt
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = 'iframe'
pio.templates.default = "plotly_white"
```

```python
df = pd.read_csv(r"C:\Users\Harshjeet\Pictures\supply_chain_data.csv")
pd.set_option('display.max_columns',None)
df.head()
```

```python
fig = px.scatter(df, x='Price', 
                 y='Revenue generated', 
                 color='Product type', 
                 hover_data=['Number of products sold'], 
                 trendline="ols")

fig.show()
```

```python
sales_data = df.groupby('Product type')['Number of products sold'].sum().reset_index()
sales_data
```

```python
pie_chart = px.pie(sales_data, values = 'Number of products sold',
                  names = 'Product type', 
                   title = 'Sales by product type',
                                      hole = 0.5
                  )
pie_chart.update_traces(textposition = 'inside', textinfo = 'percent+label')
pie_chart.show()
```

```python
total_revenue = df.groupby('Shipping carriers')['Revenue generated'].sum().reset_index()
total_revenue
```

```python
bar_chart = px.bar(total_revenue, x='Shipping carriers', y='Revenue generated', 
             title='Total Revenue by Shipping Carrier',
             labels={'Shipping carriers': 'Shipping Carrier', 'Revenue generated': 'Revenue Generated'})

# Show the bar chart
bar_chart.show()
```

```python
avg_lead_time = df.groupby('Product type')['Lead time'].mean().reset_index()
avg_lead_time
```

```python
avg_Manufacturing_costs = df.groupby('Product type')['Manufacturing costs'].mean().reset_index()
avg_Manufacturing_costs
```

```python
result = pd.merge(avg_lead_time , avg_Manufacturing_costs, on = 'Product type')
result
```

```python
result = result.rename(columns= {'Lead times':'Average Lead times', 'Manufacturing costs': 'Average Manufacturing Cost'})
                       
result                       
```

```python
revenue_chart = px.line(df, x = 'SKU', y = 'Revenue generated',
                       title = 'Revenue Generated by SKU')
revenue_chart
```

```python
bar_chart_sku = px.bar(df, x = 'SKU', y = 'Order quantities',
                      title = 'Order Quantities for each SKU'
                      )
bar_chart_sku.show()
```

```python
# Now let’s analyze the shipping cost of Carriers:

Carrer_analysis = px.bar(df, x= 'Shipping carriers', y= 'Shipping costs',
                            title= 'Shipping cost of Carriers')
Carrer_analysis.show()
```

```python
# Now let’s have a look at the cost distribution by transportation mode:

cost_distribution = px.pie(df, values = 'Costs', names = 'Transportation modes' ,
                           hole = 0.5,
                           title = 'Cost Distribution by Transportation Mode')
cost_distribution.show()
```

```python
# Let’s have a look at the average defect rate of all product types:

mean_defect_rate = df.groupby('Product type')['Defect rates'].mean().reset_index()

mean_defect_rate
```

```python
Defect_rate_analysis = px.bar(mean_defect_rate,x= 'Product type', y = 'Defect rates',
                             title = 'Avg Defect Rate by Product Type')
Defect_rate_analysis.show()
```

```python
# Now let’s have a look at the defect rates by mode of transportation

pivot_table = pd.pivot_table(df, values = 'Defect rates', index = 'Transportation modes',
                                         aggfunc = 'mean')

pivot_table
```

```python
defect_rates_transportation = px.pie(values = pivot_table['Defect rates'],
                                        names = pivot_table.index,
                                        hole = 0.5,
                                    title = 'Defect rate by Transportation')
defect_rates_transportation.show()
```
